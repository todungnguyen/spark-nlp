```python
import os
import datetime
import random
import rstr
import pandas as pd
data_path = "data/"
```

```python
def random_date(start, end):
    return start + datetime.timedelta(
        seconds = random.randint(0, int((end - start).total_seconds())),
    )

def random_dates(start, end, n):
    dates = []
    start = datetime.datetime.strptime(start, '%d/%m/%y %H:%M:%S')
    end = datetime.datetime.strptime(end, '%d/%m/%y %H:%M:%S')
    for i in range(n):
        dates.append(random_date(start, end))
    return dates

def random_ibans(n):
    ibans = []
    for i in range(n):
        ibans.append(rstr.xeger('[A-Z]{2}\d{2}?\d{4}?\d{4}?\d{4}?\d{4}?[\d]{0,2}'))
    return ibans

def random_tels(n):
    tels = []
    for i in range(n):
        tels.append(rstr.xeger('((\+)33|0)[1-9](\d{2}){4}'))
    return tels

# https://emailregex.com/
# https://codereview.stackexchange.com/questions/58269/generating-random-email-addresses
def random_emails(n):
    domains = ["hotmail.com", "gmail.com", "outlook.fr", "me.io", "yahoo.com"]
    emails = []
    for i in range(n):
        emails.append(rstr.xeger("[a-zA-Z0-9_.+-]{8,20}@") + random.choice(domains))
    return emails
```

```python
df_card = pd.read_csv(os.path.join(data_path, 'sensible/card.csv'))
```

```python
df_fname = pd.read_csv(os.path.join(data_path, 'sensible/first_name.csv'))
df_fname = df_fname.dropna()
df_fname = df_fname[df_fname['prenom'].str.contains('([a-zA-Z])\w+')]
df_fname = df_fname.reset_index(drop = True)
```

```python
df_lname = pd.read_csv(os.path.join(data_path, 'sensible/last_name.csv'))
df_lname = df_lname.dropna()
df_lname = df_lname[df_lname['nom'].str.contains('([a-zA-Z])\w+')]
df_lname = df_lname.reset_index(drop = True)
```

```python
df_adress = pd.read_csv(os.path.join(data_path, 'sensible/adress.csv'), sep=";", low_memory=False)
```

```python
df_size = min(df_card.shape[0], df_fname.shape[0], df_lname.shape[0], df_adress.shape[0])
df_size
```
